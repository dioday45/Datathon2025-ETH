**Date & Time Patterns**: 

Median hourly demand per week day

```python
# aggregated data
_ = df\
    .groupby(['hour', 'weekday'], as_index=False)\
    .agg({'demand_in_MW':'median'})

# plotting
fig = px.line(_, 
              x='hour', 
              y='demand_in_MW', 
              color='weekday', 
              title='Median Hourly Power Demand per Weekday')
fig.update_layout(xaxis_title='Hour',
                  yaxis_title='Energy Demand [MW]')
fig.show()
```

Median hourly demand per season

```python
# aggregated data
_ = df\
    .groupby(['hour', 'season'], as_index=False)\
    .agg({'demand_in_MW':'median'})

# plotting
fig = px.line(_,
              x='hour', 
              y='demand_in_MW', 
              color='season', 
              title='Median Hourly Power Demand per Season')
fig.update_layout(xaxis_title='Hour',
                  yaxis_title='Energy Demand [MW]')
fig.show()
```
